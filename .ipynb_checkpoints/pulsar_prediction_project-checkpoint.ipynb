```python
# Предоставленный датасет - это набор данных, который описывает выборку кандидатов в пульсары.

# Пульсары - это редкий тип нейтронных звезд, которые производят радиоизлучение.
# Когда пульсары вращаются, то производят обнаруживаемую картину широкополосного радиоизлучения. 
# Поскольку пульсары вращаются быстро, эта картина периодически повторяется. 
# Таким образом, поиск пульсаров включает в себя поиск периодических радиосигналов с помощью больших радиотелескопов.

# Каждый пульсар производит немного различную эмиссионную картину, которая немного меняется с каждым вращением. 
# Таким образом, потенциальное обнаружение сигнала, известное как "кандидат", усредняется по многим вращениям пульсара, определяемым длиной наблюдения. 
# В отсутствие дополнительной информации, каждый кандидат потенциально может описывать реальный пульсар. 
# Однако на практике почти все обнаружения вызваны радиочастотными помехами (RFI) и шумом, что затрудняет поиск.

# В датасете в начале перечислены переменные, а метка класса является последней записью. 
# Используемые метки классов - 0 (отрицательные) и 1 (положительные).

# Каждый кандидат описывается 8 непрерывными переменными и одной переменной класса. 
# Первые четыре представляют собой простую статистику, полученную из интегрированного профиля импульса (сложенный профиль). 
# Это массив непрерывных переменных, которые описывают разрешенную по долготе версию сигнала, которая была усредненная как по времени, так и по частоте. 
# Остальные четыре переменные получены из кривой DM-SNR.

# Среднее значение интегрированного профиля.
# Стандартное отклонение интегрированного профиля.
# Избыточный эксцесс интегрированного профиля.
# Асимметрия интегрированного профиля.
# Среднее значение кривой DM-SNR.
# Стандартное отклонение кривой DM-SNR.
# Избыточный эксцесс кривой DM-SNR.
# Асимметрия кривой DM-SNR.
# Класс объекта.
```

```python
# Для простоты обработки данных подключим необходимые библиотеки для работы с данными в датасете
import numpy as np # библиотека для математических операций
import seaborn as sns # библиотека для визуализации данных
import matplotlib.pyplot as plt # библиотека для построения графиков
import pandas as pd # библиотека для анализа данных 
```

```python
# Загрузим датасет с данными.
data = pd.read_csv("data/pulsar_stars.csv")

# посмотрим есть ли пустые данные в датасете.
print(data.isnull().sum()) 
```

```python
# Переименуем колонки, чтобы ими было проще оперировать.
data = data.rename(columns={' Mean of the integrated profile':"mean_profile",
       ' Standard deviation of the integrated profile':"std_profile",
       ' Excess kurtosis of the integrated profile':"kurtosis_profile",
       ' Skewness of the integrated profile':"skewness_profile", 
       ' Mean of the DM-SNR curve':"mean_dmsnr_curve",
       ' Standard deviation of the DM-SNR curve':"std_dmsnr_curve",
       ' Excess kurtosis of the DM-SNR curve':"kurtosis_dmsnr_curve",
       ' Skewness of the DM-SNR curve':"skewness_dmsnr_curve",
       })

# Посмотрим количество записей и тип данных в колонках.
print("data info: ", data.info())
```

```python
# Рассмотрим детально график асимметрии и эксцесса.
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.scatter(x = "kurtosis_profile",y = "skewness_profile",data=data[data["target_class"] == 1],alpha=.7,label="пульсары",s=30,color = "g",linewidths=.4,edgecolors="black")
plt.scatter(x = "kurtosis_profile",y = "skewness_profile",data=data[data["target_class"] == 0],alpha=.6,label="не пульсары",s=30,color ="r",linewidths=.4,edgecolors="black")
plt.axvline(data[data["target_class"] == 1]["kurtosis_profile"].mean(),color = "g",linestyle="dashed",label="средняя пульсара")
plt.axvline(data[data["target_class"] == 0]["kurtosis_profile"].mean(),color = "r",linestyle="dashed",label ="средняя не пульсара")
plt.axhline(data[data["target_class"] == 1]["skewness_profile"].mean(),color = "g",linestyle="dashed")
plt.axhline(data[data["target_class"] == 0]["skewness_profile"].mean(),color = "r",linestyle="dashed")
plt.legend(loc ="best")
plt.xlabel("kurtosis profile")
plt.ylabel("skewness profile")
plt.title("Точечная диаграмма асимметрии и эксцесса")
plt.subplot(122)
plt.scatter(x = "skewness_dmsnr_curve",y = 'kurtosis_dmsnr_curve',data=data[data["target_class"] == 0],alpha=.7,label="не пульсары",s=30,color ="r",linewidths=.4,edgecolors="black")
plt.scatter(x = "skewness_dmsnr_curve",y = 'kurtosis_dmsnr_curve',data=data[data["target_class"] == 1],alpha=.7,label="пульсарыet,",s=30,color = "g",linewidths=.4,edgecolors="black")
plt.axvline(data[data["target_class"] == 1]["kurtosis_dmsnr_curve"].mean(),color = "g",linestyle="dashed",label ="средняя пульсара")
plt.axvline(data[data["target_class"] == 0]["kurtosis_dmsnr_curve"].mean(),color = "r",linestyle="dashed",label ="средняя не пульсара")
plt.axhline(data[data["target_class"] == 1]["skewness_dmsnr_curve"].mean(),color = "g",linestyle="dashed")
plt.axhline(data[data["target_class"] == 0]["skewness_dmsnr_curve"].mean(),color = "r",linestyle="dashed")
plt.legend(loc ="best")
plt.xlabel("skewness_dmsnr_curve")
plt.ylabel('kurtosis_dmsnr_curve')
plt.title("Точечная диаграмма асимметрии DM-SNR и эксцесса DM-SNR")
plt.subplots_adjust(wspace =.4)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve,auc

def model(algorithm, dtrain_x, dtrain_y, dtest_x, dtest_y):
    
    # Построение математической функции на тестовых данных (https://tinyurl.com/y4aktgwa).
    algorithm.fit(dtrain_x, dtrain_y)
    predictions = algorithm.predict(dtest_x)
    
    # Посмотрим настройки алгоритма "Случайный лес".
    print(algorithm)
    
    # С помощью метода accuracy_score выведем метрику, которую можно назвать базовой. 
    # Она измеряет количество верно классифицированных объектов относительно общего количества всех объектов.
    print("\nТочность классификатора: ", accuracy_score(dtest_y, predictions))
    
    # Вывод отчета по основным метрикам. 
    print("\nОтчет по основным метрикам:\n", (classification_report(dtest_y, predictions)))
    
    # Построим матрицу ошибок (https://tinyurl.com/y2yq5y9d).
    plt.figure(figsize=(13,10))
    plt.subplot(221)
    sns.heatmap(confusion_matrix(dtest_y, predictions), annot=True, fmt="d", linecolor="k", linewidths=3)
    plt.title("МАТРИЦА ОШИБОК", fontsize=20)
    
    # Построим ROC-кривую (https://tinyurl.com/y52nymv5).
    predicting_probabilities = algorithm.predict_proba(dtest_x)[:,1]
    fpr, tpr, thresholds = roc_curve(dtest_y, predicting_probabilities)
    plt.subplot(222)
    plt.plot(fpr, tpr, label=("AUC:", auc(fpr,tpr)), color="r")
    plt.plot([1,0], [1,0], linestyle="dashed", color="k")
    plt.legend(loc="best")
    plt.title("ROC-КРИВАЯ & ПОКАЗАТЕЛЬ AUC", fontsize=20)
    
    
    # С помощью метода алгоритма feature_importances_ посмотрим какие признаки наиболее важны.
    dataframe = pd.DataFrame(algorithm.feature_importances_, dtrain_x.columns).reset_index()
    dataframe = dataframe.rename(columns={"index": "features", 0: "coefficients"})
    dataframe = dataframe.sort_values(by="coefficients", ascending = False)
    plt.subplot(223)
    ax = sns.barplot(x = "coefficients", y ="features", data=dataframe, palette="husl")
    plt.title("ВАЖНОСТЬ ПИЗНАКОВ", fontsize =20)
    for i, j in enumerate(dataframe["coefficients"]):
        ax.text(.011, i , j, weight = "bold")
    
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# С помощью библиотеки (sklearn) для машинного обучения попробуем научиться классифицировать объекты.
y = data.target_class # класс объекта выбран за y
x_data = data.drop(["target_class"],axis = 1) 
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)).values # нормализация данных по x

# Приготовим данные для машинного обучения.
train_X, test_X, train_Y, test_Y = train_test_split(x, y, test_size = 0.3, random_state = 42)

# Применим алгоритм "Случайного леса" (https://tinyurl.com/ycc8o59r).
rf = RandomForestClassifier()
model(rf, train_X, train_Y, test_X, test_Y)
```

```python
# Попробуем классифицировать объекты без использования сторонней библиотеки машинного обучения.
x_data = data.drop(["target_class"], axis = 1)
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
y = data.target_class.values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
# Метод инициализирующий веса (https://tinyurl.com/y5wha9vh). 
def initializeWeightBias(dimension):
    w = np.full((dimension,1), 0.01)
    b = 0.0
    return w, b

# Функция активации (https://tinyurl.com/y4tfdpjg) Сигмоида.
def sigmoidFunc(z):
    y_head = 1/(1 + np.exp(-z))
    return y_head

# Метод распространения ошибки (https://tinyurl.com/y8vyrpcp, https://tinyurl.com/y4pdng44).
def fwPropagation(w, b, x_train, y_train):
    #прямое
    z = np.dot(w.T, x_train.T) + b
    y_head = sigmoidFunc(z)
    loss = -y_train*np.log(y_head)-(1-y_train)*np.log(1-y_head)
    cost = (np.sum(loss))/x_train.shape[0]
    
    #обратное
    deriWeight = (np.dot(x_train.T,((y_head-y_train.T).T)))/x_train.shape[0] 
    deriBias = np.sum(y_head-y_train.T)/x_train.shape[0]
    graDescents = {"deriWeight" : deriWeight, "deriBias" : deriBias}
    
    return graDescents, cost
    
# Метод обновления параметров.    
def update(iterNumber, w, b, x_train, y_train, learningRate):
    costList = []
    index = []
    for i in range(iterNumber + 1):
        graDescents, cost = fwPropagation(w, b, x_train, y_train)
        w = w - learningRate*graDescents["deriWeight"]
        b = b - learningRate*graDescents["deriBias"]
        
        if i % 10 == 0:
            costList.append(cost)
            index.append(i)
            print("\nЗначение функции затрат после {} итераций = {}".format(i, cost))
            
    parameters = {"weight" : w, "bias" : b}
    
    return parameters, costList, index     
 
def plotGraph(index, costList):
    plt.plot(index, costList)
    plt.ylabel("Cost")
    plt.show()
    
def predict(w, b, x_test):
    z = np.dot(w.T, x_test.T) + b
    y_head = sigmoidFunc(z)
    yPrediction = np.zeros((1,x_test.shape[0]))
    
    for i in range(y_head.shape[1]):
        if y_head[0,i] <= 0.5:
            yPrediction[0,i] = 0
        else:
            yPrediction[0,i] = 1
            
    return yPrediction

# Построение классификатора методом логистической регрессии (https://tinyurl.com/y6omhmux).
def logisticRegression(x_train, y_train, x_test, y_test, iterNumber, learningRate):
    dimension = x_train.shape[1]
    
    # Инициализируем веса.
    w, b = initializeWeightBias(dimension)
    
    parameters, costList, index = update(iterNumber, w, b, x_train, y_train, learningRate)
    
    predictionTest = predict(parameters["weight"], parameters["bias"], x_test)
    
    print("\nТочность классификатора: {} %".format(100 - np.mean(np.abs(predictionTest - y_test)) * 100))
    
    plotGraph(index, costList)
    
logisticRegression(x_train, y_train, x_test, y_test, iterNumber = 30, learningRate = 0.5)
```
